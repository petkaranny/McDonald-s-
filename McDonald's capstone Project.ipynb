 # <font color = darpink> 2.4 OUTLIER TREATMENT </font>
 

```python
col=df.select_dtypes(exclude='object').columns
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[col] < (Q1 - 1.5*IQR))| (df[col] > (Q3 + 1.5*IQR))).any(axis=1)]
```
